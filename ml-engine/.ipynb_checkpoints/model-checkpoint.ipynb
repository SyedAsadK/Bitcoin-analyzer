```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import root_mean_squared_error
from statsmodels.tsa.arima.model import ARIMA
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

plt.style.use('ggplot')
```

```python
# Load the specific BTC dataset
df = pd.read_csv('BTC.csv')

# We only need the Date and Close price columns for our models
df = df[['Date', 'Close']]

# Convert the string dates (MM/DD/YY) into actual pandas datetime objects
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%y')
# CRITICAL FIX: Sort the dates chronologically (oldest first, newest last)
df = df.sort_values('Date').reset_index(drop=True)

# Set Date as the index for plotting and time-series logic
df.set_index('Date', inplace=True)

# ---> ADD THIS LINE <---
# Explicitly set the frequency to Daily ('D') and forward-fill any missing days in the Kaggle dataset
df = df.asfreq('D', method='ffill')
# Plot the raw data for Chapter 5 of your report
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], color='blue')
plt.title('Bitcoin Historical Closing Prices')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.show()

# Keep the last 100 days for testing, train on the rest
test_size = 100
train_data = df[:-test_size]
test_data = df[-test_size:]

print(f"Training data size: {len(train_data)} days")
print(f"Testing data size: {len(test_data)} days")
```

```python
print("--- Training Random Forest ---")

# Create a supervised learning dataset (using lag features)
def create_rf_features(data, lookback=7):
    X, y = [], []
    prices = data['Close'].values
    for i in range(len(prices) - lookback):
        X.append(prices[i:(i + lookback)])
        y.append(prices[i + lookback])
    return np.array(X), np.array(y)

X_train_rf, y_train_rf = create_rf_features(train_data)
X_test_rf, y_test_rf = create_rf_features(pd.concat([train_data.iloc[-7:], test_data]))

# Train the model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train_rf, y_train_rf)

# Predict
rf_predictions = rf_model.predict(X_test_rf)
rf_rmse = root_mean_squared_error(test_data['Close'], rf_predictions)
print(f"Random Forest RMSE: {rf_rmse:.2f}")
```

```python
print("--- Training ARIMA (Rolling Forecast) ---")

# We create a list of history that will grow day by day
history = [x for x in train_data['Close']]
test_actuals = test_data['Close'].values
arima_predictions = []

# Loop through each day in the test set
for t in range(len(test_actuals)):
    # Create and fit the model on the current history
    model = ARIMA(history, order=(5, 1, 0))
    # We suppress warnings inside the loop to keep the output clean
    import warnings
    from statsmodels.tools.sm_exceptions import ConvergenceWarning
    warnings.simplefilter('ignore', ConvergenceWarning)
    
    model_fit = model.fit()
    
    # Predict exactly ONE step into the future
    yhat = model_fit.forecast()[0]
    arima_predictions.append(yhat)
    
    # Add the ACTUAL real-world price to the history for the next loop's context
    history.append(test_actuals[t])

# Calculate new Error
arima_rmse = root_mean_squared_error(test_actuals, arima_predictions)
print(f"ARIMA Rolling RMSE: {arima_rmse:.2f}")
```

```python
print("--- Training LSTM ---")

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_train = scaler.fit_transform(train_data)
scaled_test = scaler.transform(test_data)

# Create sequences for LSTM (shape: [samples, time steps, features])
def create_lstm_sequences(data, lookback=60):
    X, y = [], []
    for i in range(lookback, len(data)):
        X.append(data[i-lookback:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

lookback = 60
# We append the end of train_data to test_data to get the first 60 days of context for testing
lstm_test_inputs = np.vstack((scaled_train[-lookback:], scaled_test))

X_train_lstm, y_train_lstm = create_lstm_sequences(scaled_train, lookback)
X_test_lstm, y_test_lstm = create_lstm_sequences(lstm_test_inputs, lookback)

# Reshape for Keras
X_train_lstm = np.reshape(X_train_lstm, (X_train_lstm.shape[0], X_train_lstm.shape[1], 1))
X_test_lstm = np.reshape(X_test_lstm, (X_test_lstm.shape[0], X_test_lstm.shape[1], 1))

# Build and compile the network architecture
lstm_model = Sequential()
lstm_model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train_lstm.shape[1], 1)))
lstm_model.add(LSTM(units=50))
lstm_model.add(Dense(units=1))

lstm_model.compile(optimizer='adam', loss='mean_squared_error')
lstm_model.fit(X_train_lstm, y_train_lstm, epochs=10, batch_size=32, verbose=1)

# Predict and inverse transform back to USD prices
lstm_preds_scaled = lstm_model.predict(X_test_lstm)
lstm_predictions = scaler.inverse_transform(lstm_preds_scaled)

lstm_rmse = root_mean_squared_error(test_data['Close'], lstm_predictions)
print(f"LSTM RMSE: {lstm_rmse:.2f}")
```

```python
plt.figure(figsize=(14, 7))

# Plot Actual Prices
plt.plot(test_data.index, test_data['Close'], label='Actual Bitcoin Price', color='black', linewidth=2)

# Plot Predictions
plt.plot(test_data.index, arima_predictions, label=f'ARIMA (RMSE: {arima_rmse:.0f})', color='orange')
plt.plot(test_data.index, rf_predictions, label=f'Random Forest (RMSE: {rf_rmse:.0f})', color='green')
plt.plot(test_data.index, lstm_predictions, label=f'LSTM (RMSE: {lstm_rmse:.0f})', color='purple')

plt.title('Bitcoin Price Prediction Comparison (Last 100 Days)')
plt.xlabel('Date')
plt.ylabel('Price in USD')
plt.legend()
plt.show()
```
